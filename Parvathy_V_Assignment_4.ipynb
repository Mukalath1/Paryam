```python
data['Calories'].replace([np.nan], data['Calories'].mean(), inplace=True)
data['Date'] = data['Date'].fillna(data['Date'].mode()[0])
data
```

```python
sns.scatterplot(x=data['Maxpulse'],y=data['Calories'],hue=data['Date'])
plt.title('MAXPULSE Vs CALORIES')
plt.show()
```
